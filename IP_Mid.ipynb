```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_bgr = cv2.imread('/content/eren.jpg')
```

# **Answer to Question no 01**

1. loading the image in grayscale image

```python
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
```

2. Using 7x7 Kernel size, applying gaussian filter to smoothing the image

```python
smoothing_karnel = (7, 7)
low_pass = cv2.GaussianBlur(image_gray, smoothing_karnel, 0)
```

3. Displaying both Original and Blurring image uisng matplotlib

```python
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(image_gray, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(low_pass, cmap='gray')
plt.title('Blurred Image')
```

# **Answer to Question no 02**

- High_pass = Original - low_pass

```python
high_pass = image_gray - low_pass
plt.figure(figsize=(10, 10))
plt.imshow(high_pass, cmap='gray')
plt.title('High Pass Filter')
```

- Original = low_pass + high_pass

```python
Original_image = low_pass + high_pass
plt.figure(figsize=(10, 10))
plt.imshow(Original_image, cmap='gray')
plt.title('Original Image')
```

# **Answer to Question no 03**

-  Histogram of Original  Gray scale Image

```python
hist = cv2.calcHist([image_gray], [0], None, [256], [0, 256])
```

- Equalizing the gray scale iamge

```python
equalized_histogram = cv2.equalizeHist(image_gray)
```

- Displaying Both

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(hist)
plt.title('Original Image Histogram')
plt.subplot(1, 2, 2)
plt.plot(cv2.calcHist([equalized_histogram], [0], None, [256], [0, 256]))
plt.title('Histogram Equalization')
```
