# Segmentation

1. Global and Adaptive threshholding Side by side



```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

```

```python
img_bgr = cv2.imread('/content/eren.jpg')

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
def adaptive_threshold(img):
  img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  img_gray = cv2.medianBlur(img_gray, 3)
  img_adaptive = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 5, 2)
  return img_adaptive

def global_threshold(img):
  img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  img_gray = cv2.medianBlur(img_gray, 3)
  ret, img_global = cv2.threshold(img_gray, 140, 255, cv2.THRESH_BINARY)
  return img_global


plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.title('Global Thresholding')
plt.imshow(global_threshold(img_rgb), cmap='gray')
plt.subplot(1, 2, 2)
plt.title('Adaptive Thresholding')
plt.imshow(adaptive_threshold(img_rgb), cmap='gray')

plt.show()



```

2. Canny Edge detection

```python
def canny_edge(img):
  img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  img_gray = cv2.medianBlur(img_gray, 3)
  img_canny = cv2.Canny(img_gray, 100, 170)
  return img_canny

plt.figure(figsize=(6, 6))
plt.title('Canny Edge Detection')
plt.imshow(canny_edge(img_rgb), cmap='gray')
plt.show()
```
