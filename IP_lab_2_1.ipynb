```python
import cv2 # Open CV Load
from PIL import Image # Pillow load
from matplotlib import pyplot as plt # matpotlib load from graph
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
image_path = '/content/image.jpg'
```

```python
# loading using CV
image_bgr = cv2.imread(image_path)
```

Loading BGR Image

```python
plt.imshow(image_bgr)
```

Converting into Gray

```python
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
gray = plt.imshow(image_gray, cmap='gray')
```

# **Global**
<br>
thresholding manual or user defined 2 tai hote pare
minimum , maximum


- Apply Binary Thresholding

```python
# Binary Image (Thresholding)
# 120 - thresh value, 255 - max value , thresh type.
binary = cv2.threshold(image_gray, 120, 255, cv2.THRESH_BINARY)
plt.imshow(binary[1], cmap='gray')
```

# **Adaptive**

-  Dual(Gaussian)

```python
# 255 - max value, adaptive type, threshold type, blocksize, c value
dual_threshold = cv2.adaptiveThreshold(image_gray, 225, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
plt.imshow(dual_threshold, cmap='gray')
```

- Dual(mean)

```python
mean = cv2.adaptiveThreshold(image_gray, 225, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
plt.imshow(mean, cmap='gray')
```

# **Contrust Stretching**

```python
min_val, max_val = np.min(image_gray), np.max(image_gray)
contrust_stretched = (255*((image_gray - min_val) / (max_val - min_val))).astype(np.uint8)
plt.imshow(contrust_stretched, cmap='gray')
```

# **Negative, Brightness and Darkness**

- Negative

```python
negative = 255 - image_gray
plt.imshow(negative, cmap='gray')
```

- Brightness

```python
brightness = cv2.convertScaleAbs(image_gray, alpha=1.5, beta=50) # alpha = contrast , beta = brightness
plt.imshow(brightness, cmap='gray')
```

- Darkness

```python
darkness = cv2.convertScaleAbs(image_gray, alpha=0.7, beta=-50)
plt.imshow(darkness, cmap='gray')
```
